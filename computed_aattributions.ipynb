### Performing 1 sample t-test

```python
header = all_df.columns.tolist()
```

```python
from scipy.stats import ttest_1samp
```

```python
tval,pval = ttest_1samp(all_df.values.tolist(),0,0)
```

```python
[header[i] for i in range(len(pval)) if pval[i]>0.05]
```

### Two sample ttest

```python
from scipy.stats import ttest_ind
```

```python
tval , pval = ttest_ind(female_df.values.tolist(),male_df.values.tolist())
```

```python
[header[i] for i in range(len(pval)) if pval[i]<0.05]
```

```python
high_memory = all_df[all_df['memory'] == 1]
low_memory = all_df[all_df['memory'] == 0]
```

```python
tval , pval = ttest_ind(low_memory.values.tolist(),high_memory.values.tolist())
```

```python
[header[i] for i in range(len(pval)) if pval[i]<0.05]
```
